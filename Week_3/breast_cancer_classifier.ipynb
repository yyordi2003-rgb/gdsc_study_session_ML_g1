```python
#column 'Bare Nuclei' has a value of '?' which needs to be filled with another number value
#there are 16 '?' values in 'Bare Nuclei' column
```

```python
#lets fill '?' with a nan
df['Bare Nuclei']= df['Bare Nuclei'].replace('?',np.nan)
df[df['Bare Nuclei']=='?'].shape[0]
```

```python
#now lets replace the nan value with median of the column 'Bare Nuclei'
# First, convert the column to a numeric type
df['Bare Nuclei'] = pd.to_numeric(df['Bare Nuclei'])
df['Bare Nuclei']= df['Bare Nuclei'].fillna(df['Bare Nuclei'].median())
df['Bare Nuclei'].unique()
```

```python
#lets drop non-informative columns, for example column 'Sample code number', since it has
# a high number of unqiue values
df.drop('Sample code number', axis=1, inplace=True)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

#lets separate the features and the targets
x= df.drop('Class', axis=1)
y= df['Class']
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3)
```

```python
#when we use scaling we must use it after splitting the dataset into trianing and test set, to prevent data leakge
#for scaling i will use Standardization
#it is a method of scaling data where the values are transformed to have a mean of 0 and a standard deviation of 1
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

sc= StandardScaler()
x_train_scaled= sc.fit_transform(x_train)
x_test_scaled= sc.transform(x_test)
```

```python
#lets train and test a logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

lr= LogisticRegression().fit(x_train_scaled, y_train)
y_pred= lr.predict(x_test_scaled)
y_pred
```

```python
accuracy= accuracy_score(y_test, y_pred)
accuracy
```

```python
cm= confusion_matrix(y_test, y_pred)
cm
```

```python
cf= classification_report(y_test, y_pred)
print(cf)
```

```python
#the model has a high accuracy score, which makes our predictions accurate
#based on the confusion matrix our model can predict fully accurate of one class (positive class), while having a slightly less inaccuracy when predicting the other class(negative class)
#our f1-score is high implying that Perfect Performance. The model has perfect precision and recall.
#overall our model has a good prediction result, but it might have better prediciton power with some feature engineering and other model error minimizations method
```
